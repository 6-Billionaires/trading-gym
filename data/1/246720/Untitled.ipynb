```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = "0-246720-20180406-order.csv"
```

```python
df = pd.read_csv(path, index_col = 0)
```

```python
df.loc[:,'SellHoga1'].head()
```

```python
price_list = list(df.loc[:,'SellHoga1'].values)
rtn_list = []

for i in range(len(price_list)):    
    time = df.index[i]
    base_price = price_list[i]
    max_price = np.array(price_list[i:i+60]).max()
    min_price = np.array(price_list[i:i+60]).min()
    rtn = (max_price+100) / (base_price+100) * 100 - 100
    rtn_list.append(rtn)
    print(time, base_price,min_price,max_price,rtn)
    
```

```python
rtn_list = rtn_list[:3600]
```
